```python
import pandas as pd
df1 = pd.read_csv("Food_Supply_Quantity_kg_Data.csv")
df2 = pd.read_csv("Deaths_csv_cleaned.csv")
```

```python
# Deaths_csv_cleaned.csv had the Country column as "location". Renaming the column to Country to fit the Food_Supply .csv

df2_renamed = df2.rename(columns={'location': 'Country'}, inplace=False)
df2_renamed.head()
```

```python
df1.head()
```

```python
merged_df = df2_renamed.merge(df1, how="inner", on = ['Country'])
merged_df.head()
```

```python
len(df1.index), len(df2.index), len(df2_renamed.index), len(merged_df.index)
```

```python
merged_df.to_csv("Merged.csv")
```
